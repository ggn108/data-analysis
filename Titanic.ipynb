# Titanic Data Analysis

The RMS Titanic sank in the North Atlantic in 1912 after hitting a iceburg.  
Around two-thirds of the people aboard died in the tragedy.  

## 1. Research questions
What was the difference in survival between passengers of different ticket classes?

```python
import pandas as pd
```

```python
titanic = pd.read_csv('data/titanic.csv')
titanic
```

We have 891 passengers and 15 features.  

```python
titanic.describe()
```

## 2. Analysis - Investigate the relationship between survived, sex and passenger class (pclass). 

```python
table = titanic.groupby(['sex','pclass','survived'])['survived'].aggregate('count').unstack()
table
```

```python
table.plot(kind='bar',stacked=True);
```

Generally females survived more than males.  
Female from first class were luckiest.  
Males from third class were most unlucky.

The evacuation seems to have started with the upper class ladies, but then they probably ran out of life boats or the ship sank before the rescue operation finished.
